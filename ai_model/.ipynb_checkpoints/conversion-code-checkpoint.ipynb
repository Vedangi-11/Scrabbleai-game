```python
# --- PATCH OPTIONAL DEPENDENCIES SAFELY ---
import sys, types

# Create a dummy jax2tf namespace with a fake function to satisfy import
jax2tf_module = types.SimpleNamespace()
jax2tf_module.convert = lambda *args, **kwargs: None  # mock convert()

# Now fake all required optional packages
fake_modules = {
    "tensorflow_decision_forests": types.SimpleNamespace(),
    "tensorflow_hub": types.SimpleNamespace(),
    "flax": types.SimpleNamespace(),
    "jax": types.SimpleNamespace(experimental=types.SimpleNamespace(jax2tf=jax2tf_module)),
    "jaxlib": types.SimpleNamespace(),
    "jax.experimental": types.SimpleNamespace(jax2tf=jax2tf_module),
    "jax.experimental.jax2tf": jax2tf_module,
}

sys.modules.update(fake_modules)

# --- IMPORTS ---
import os
import tensorflow as tf
import tensorflowjs as tfjs

print("✅ TensorFlow version:", tf.__version__)
print("✅ TensorFlow.js version:", tfjs.__version__)

# --- CONVERSION ---
os.makedirs("tfjs_scrabble_model", exist_ok=True)
model = tf.keras.models.load_model("model/model_weights_lstm.h5")

tfjs.converters.save_keras_model(model, "tfjs_scrabble_model")

print("✅ Conversion successful! Files generated in tfjs_scrabble_model/:")
print(os.listdir("tfjs_scrabble_model"))
```
